```python
import image_transformation_functions as transf_func
import os
import pandas as pd
import numpy as np
from PIL import Image, ImageDraw, ImageOps
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# save class labels from directory
class_labels = transf_func.save_class_labels_from_directory("C:/Users/jamie/Python Projects/Dissertation/Task/datasets/WSP-8 No augmentation/all labels/")

# create df of labels
class_names = {0:'Articulated', 1:'Bus', 2:'Car', 3:'Coach', 4:'LGV', 5:'Rigid 2 Axle', 6:'Rigid 3 Axle', 7:'Rigid 4 Axle', 8:'Taxi'}

classes_df = transf_func.create_class_labels_df(class_labels, class_names)
classes_df["class_label_txt"].value_counts()
```

```python
# create df for frequency counts of all vehicles
freq_counts = classes_df["class_label_txt"].value_counts().reset_index(name='Frequency').rename(columns={'class_label_txt':'Vehicle Type'})
```

```python
# bar plot of class balance in expanded dataset
sns.set_palette("deep")
ax = sns.barplot(x=freq_counts["Vehicle Type"], y=freq_counts["Frequency"])
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_title("Class Balance in Expanded WSP Dataset")
plt.savefig("C:/Users/jamie/Python Projects/Dissertation/Task/EDA Code/Original WSP Dataset Charts/second_class_balance.png", bbox_inches="tight")
plt.show()
```
